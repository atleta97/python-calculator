```python
def calculatorProg():
    while True:
        print("""
Welcome to my simple calculator program
This program will allow you to perform basic arithmetic operations 
""")
        
        userInput = input(
            "Type in 'Y' to continue or any other key to exit:"
        )

        if userInput.strip().lower() ==  "y":
            print("Let's go!")
            

            while True:
                try:
                    num1 = float(input("Please input your first number:"))
                    break
                except ValueError:
                    print("Please enter a valid number")
                    continue
            
            while True:
                operator = input(
                    "Please choose your operator method:\n"
                    "Press + for addition\n"
                    "Press - for Subtraction\n"
                    "Press * for multiplication\n"
                    "Press / for division\n"
                )
                if operator in ["+", "-", "*", "/"]:
                    break
                print("Please input a valid operator"
                )
                    

            while True:
                try:
                    num2 = float(input("Please input your second number:"))
                    break
                except ValueError:
                    print("Please enter a valid number")
                    continue
            
            match operator:    
                case "+":
                    result = num1 + num2

                case "-":
                    result = num1 - num2

                case "*":
                    result = num1 * num2

                case "/":
                    while num2 == 0:
                        print("Cannot divide by zero")
                        num2 = float(
                            input(
                                "Please re-enter a non-zero number: "
                            )
                        )

                    result = num1 / num2

                case _:
                    print("Invalid Operator")
                    continue

            print(f"The result is: {num1} {operator} {num2} = {result}")
            
            again = input(
                "Do you want to carry out another operation (Y/N):"
            )

            if again.strip().lower() != 'y':
                break

        else:
            break
    
    print("Bye!)
    
    
calculatorProg()
```
